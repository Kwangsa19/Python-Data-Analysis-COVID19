### Task 3.3: find maximum death rate for all of the countries. 

```python


# Assuming your date columns are like "4/22/20", "4/24/20", etc.
# Replace it with the actual date columns in your dataset
date_columns = corona_dataset_aggregated.columns[1:]  # Assuming the first column is not a date

# Get a list of countries
countries = list(corona_dataset_aggregated.index)

# Initialize a list to store max death rates
max_death_rates = []

for c in countries:
    country_data = corona_dataset_aggregated.loc[c, date_columns]
    
    # Check if the country_data has sufficient data
    if len(country_data) < 2 or country_data.isnull().any():
        max_death_rates.append(None)  # or any suitable placeholder
        continue
    
    # Calculate the maximum death rate
    max_death_rate = country_data.diff().max()
    
    max_death_rates.append(max_death_rate)

# Add the "max_death_rate" column to the DataFrame
corona_dataset_aggregated["max_death_rate"] = max_death_rates


#corona_dataset_aggregated.dtypes
#max_death_rates = corona_dataset_aggregated.apply(
#    lambda x: (pd.to_numeric(x.iloc[1:], errors='coerce') - pd.to_numeric(x.iloc[:-1], errors='coerce')).max(),
#    axis=1
#)

# Add the "max_death_rate" column
#corona_dataset_aggregated["max_death_rate"] = max_death_rates




# Calculate maximum daily difference (assuming numerical columns)
# max_death_rates = corona_dataset_aggregated.apply(
#    lambda x: (x.iloc[1:] - x.iloc[:-1]).max(), axis=1
#)

# Add the "max_death_rate" column
#corona_dataset_aggregated["max_death_rate"] = max_death_rates
```

```python
# Find the correlation 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Define x and y using your actual data
x = data["Social support"]
y = data["max_death_rate"]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Create a scatter plot
sns.scatterplot(x=x, y=y, label='Data Points')

# Add the linear regression line to the plot
sns.lineplot(x=x, y=(slope * x + intercept), color='red', label=f'Linear Regression\nR={r_value:.4f}, R^2={r_value**2:.4f}')

# Add labels and title
plt.xlabel('Social support')
plt.ylabel('Max Death Rate')
plt.title('Scatter Plot with Linear Regression: Social support vs Max Death Rate')
plt.legend()

# Show the plot
plt.show()

# Display regression results
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R-value: {r_value:.4f}")
print(f"R^2: {r_value**2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Standard error: {std_err:.4f}")

# Similarly, for R^2:
# If R^2 is close to 1, it suggests a strong correlation.
# If R^2 is around 0.5, it suggests a moderate correlation.
# If R^2 is close to 0, it indicates a weak correlation.
```
